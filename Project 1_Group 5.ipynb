```python
#To store the error rate(1-Acccuracy Rate)
error_rate = []
#Iterates through K values from 1 to 39.
#Creates a new KNN classifier object for each K value.

for i in range(1,40):
 knn = KNeighborsClassifier(n_neighbors=i)
 knn.fit(x_train,y_train)
 pred_i = knn.predict(x_test)
 #print (pred_i)
 #print (1-accuracy_score(Y_test, pred_i))
 error_rate.append(1-accuracy_score(y_test, pred_i))

#Plots the error rate for each K value against the corresponding K value (1 to 39). The plot uses blue color, dashed line style, and red circular markers for better visualization.
plt.figure(figsize=(10,6))
plt.plot(range(1,40),error_rate,color='blue', linestyle='dashed', 
         marker='o',markerfacecolor='red', markersize=10)
plt.title('Error Rate vs. K Value') #Sets the  title of the plot.
plt.xlabel('K') #Labels the x-axis as "K"
plt.ylabel('Error Rate') #Labels the y-axis as "Error Rate"
plt.show() #Displays the Generated Plot.
print("Minimum error:-",min(error_rate),"at K =",error_rate.index(min(error_rate))+1)
```
